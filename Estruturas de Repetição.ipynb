#### 1. Faça um programa que peça uma nota, entre zero e dez. Mostre uma mensagem caso o valor seja inválido e continue pedindo até que o usuário informe um valor válido.

````python
nota = int(input('Digite uma nota de 0 a 10'))
while nota < 0 or nota > 10:
  nota = int(input("Digite uma nota de 0 a 10"))
  if nota < 0 or nota > 10:
    print("Digite uma nota de 0 a 10")
print(nota)
````

#### 2. Faça um programa que leia um nome de usuário e a sua senha e não aceite a senha igual ao nome do usuário, mostrando uma mensagem de erro e voltando a pedir as informações.

````python
usuario = input('digite seu usuario: ')
senha = input('digite sua senha: ')

while usuario == senha:
  if usuario == senha:
    print('usuario e senha nao pode ser iguais')
    usuario = input('digite seu usuario: ')
    senha = input('digite sua senha: ')
  if usuario != senha:
    print('usuario logado')    
print(usuario,senha)
````

#### 3. Faça um programa que leia e valide as seguintes informações (e para cada uma delas, continue pedindo a informação até o usuário inserir corretamente):
##### Nome: maior que 3 caracteres;
##### Idade: entre 0 e 150;
##### Salário: maior que zero;
##### Sexo: 'f' ou 'm';
##### Estado Civil: 's', 'c', 'v', 'd';

````python


while True:
    nome = input("Digite seu nome (maior que 3 caracteres): ")
    if len(nome) > 3:
        break

while True:
    idade = int(input("Digite sua idade (entre 0 e 150): "))
    if idade >= 0 and idade <= 150:
        break

while True:
    salario = float(input("Digite seu salário (maior que zero): "))
    if salario > 0:
        break

while True:
    sexo = input("Digite seu sexo ('f' ou 'm'): ")
    if sexo == 'f' or sexo == 'm':
        break

lista = ['s', 'c', 'v', 'd']
estado = ''
while  estado not in lista:
    estado = input("Digite seu estado civil ('s', 'c', 'v', 'd'): ")
    
print(f'''Dados cadastrados:
Nome: {nome}
Idade: {idade}
Salário: {salario}
Sexo: {sexo}
Estado civil: {estado}
''')
````

Supondo que a população de um país A seja da ordem de 80000 habitantes com uma taxa anual de crescimento de 3% e que a população de B seja 200000 habitantes com uma taxa de crescimento de 1.5%. Faça um programa que calcule e escreva o número de anos necessários para que a população do país A ultrapasse ou iguale a população do país B, mantidas as taxas de crescimento.

````python
populacao_A = 80000
populacao_B = 200000
taxaAnualA = 0.03
taxaAnualB = 0.015

anos = 0
while populacao_A < populacao_B:
  populacao_A += populacao_A * taxaAnualA
  populacao_B += populacao_B * taxaAnualB
  anos += 1
print(f'depois de {anos} a populacao A vai passar a populacao B')
````

#### 6. Faça um programa que peça para o usuário inserir o faturamento dos últimos 5 meses (individualmente) e informe o maior faturamento

````python
faturamentos = []

for i in range(5):
  while True:
    faturamento = float(input(f'Digite o faturamento do mes {i + 1}'))
    faturamentos.append(faturamento)
    break

maximo_fat = max(faturamentos)
print(maximo_fat)
````

#### 7. Faça um programa que peça para o usuário inserir o faturamento dos últimos 5 meses (individualmente) e informe o faturamento total (soma) e o faturamento médio por mês (média).
